Pareto Distribution

```python
sales_amount=abc_analysis(df1['Sales Amount'],True)
```

```python
a_index=sales_amount['Aind']
b_index=sales_amount['Bind']
c_index=sales_amount['Cind']


cond_list=[df1.index.isin(a_index),
          df1.index.isin(b_index),
          df1.index.isin(c_index)]


choice_list=['A','B','C']
df1['sales_amount']=np.select(cond_list,choice_list)
df1.sort_values(by=['Sales Amount'],ascending=False)
```

```python
df1['sales_amount']
```

```python
Profit=abc_analysis(df1['Sales Margin Amount'],True)
```

```python
a_index=Profit['Aind']
b_index=Profit['Bind']
c_index=Profit['Cind']


cond_list=[df1.index.isin(a_index),
          df1.index.isin(b_index),
          df1.index.isin(c_index)]


choice_list=['A','B','C']
df1['Profit']=np.select(cond_list,choice_list)
df1.sort_values(by=['Sales Margin Amount'],ascending=False)
```
